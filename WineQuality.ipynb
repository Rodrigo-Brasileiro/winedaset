# Passo 0: Imports
## Todas as bibliotecas necessárias para realizar a atividade

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

# Passo 1: Declarar o dataset na variável

```python
data = pd.read_csv('winequality-white.csv', delimiter=';')
```

# Passo 2: Selecionando as características para o modelo de regressão linear

```python
propriedades = data[['quality','volatile acidity', 'alcohol', 'free sulfur dioxide']]
target = data['total sulfur dioxide']
```

<h1>Passo 3: Dividir os dados em conjuntos de treinamento e teste</h1>

```python
X_train, X_test, y_train, y_test = train_test_split(propriedades, target, test_size=0.2, random_state=42)
```

<h1> Passo 4: Criar o modelo de regressão linear </h1>

```python
modelo = LinearRegression()
```

<h1> Passo 5: Treinar o modelo </h1>

```python
modelo.fit(X_train, y_train)
```

<h1> Passo 6: Fazer previsões no conjunto de teste </h1>

```python
y_pred = modelo.predict(X_test)
```

<h1> Passo 7: Avaliar o modelo </h1>

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f'Mean Squared Error: {mse}')
print(f'R^2 Score: {r2}')
```

<h1> Passo 8: Plotar a regressão linear </h1>

```python
plt.scatter(y_test, y_pred)
plt.xlabel('Valores Reais')
plt.ylabel('Previsões')
plt.title('Regressão Linear - Qualidade do Vinho')
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red')  # Linha de 45 graus para referência
plt.show()
```
